```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
ev_df = pd.read_csv("Electric_Vehicle_Population_Data.csv")
city_df = pd.read_csv("us_cities.csv")
```

```python
ev_df
```

```python
city_df
```

```python
#merge two dataframe based on city

merged_df = pd.merge(ev_df, city_df, on='City')

# print merged DataFrame
print(merged_df)
```

```python
#Extracting features based on common knowledge
column_headers = merged_df.columns.tolist()
column_headers
```

```python
car_num['count'] = merged_df.groupby('City')['VIN (1-10)'].count()
car_num
```

```python
car_num['AvgIncome'] = merged_df.groupby('City')['AvgIncome'].mean()
car_num
```

```python
df = car_num[['count','AvgIncome']]
df['count'].max()
```

```python
df.describe()
```

```python
df['count']
```

```python
# plots of the data distributions
plt.figure(figsize=(30, 30))

# Histogram for Sales
plt.subplot(2, 1, 1)
sns.histplot(x=df.index,y=df['count'], bins=20, color='skyblue', edgecolor='black',kde=True)
plt.title('Count Distribution')
plt.xlabel('City')
plt.ylabel('Count')

# Histogram for Temperature
plt.subplot(2, 1, 2)
sns.histplot(x=df.index,y=df['AvgIncome'], bins=20, color='salmon', edgecolor='black',kde=True)
plt.title('AvgIncome Distribution')
plt.xlabel('City')
plt.ylabel('AvgIncome')

plt.tight_layout()
plt.show()
```

```python
correlation = df['AvgIncome'].corr(df['count'])

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['AvgIncome'], df['count'], color='blue', alpha=0.7)
plt.title('Relationship between AvgIncome and Sales count')
plt.xlabel('AvgIncome')
plt.ylabel('count')
plt.grid(True)
plt.show()

# Print the correlation coefficient
print(f'Correlation Coefficient between AvgIncome and count: {correlation}')
```

```python
# Define the independent variable (X) and the dependent variable (y)
X = df['AvgIncome']
X = sm.add_constant(X)  # Add a constant term to the independent variable
y = df['count']

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the regression summary
print(model.summary())
```
